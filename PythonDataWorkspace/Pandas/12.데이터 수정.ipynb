# 12. 데이터 수정

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

## Column 수정

```python
df['학교'].replace({'북산고':'상북고','능남고':'무슨고'}) # 북산고 → 상북고, 능남고 → 무슨고
# Inplace 필요
```

```python
df
```

```python
df['학교'].replace({'북산고':'상북고'}, inplace=True)
df
```

```python
df['SW특기'].str.lower() # SW특기 Series의 문자를 소문자로
```

```python
df['SW특기'] = df['SW특기'].str.lower() # 기존 데이터에 적용
df
```

```python
df['SW특기'] = df['SW특기'].str.upper()
df
```

```python
df['학교'] = df['학교'] + '등학교' # 학교 데이터 + 등학교
df
```

```python
# df['학교'] += '입니다' # 이와 같이 적용가능
# df
```

## Column 추가

```python
df['총합'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회'] # 기존에 없던 column명이 변수로 들어오면 우측에 추가됨
df
```

```python
df['결과'] = 'Fail' # 결과 컬럼을 추가하고 전체 데이터는 Fail로 초기화
df
```

```python
df.loc[df['총합'] > 400, '결과'] = 'Pass' # 총합이 400보다 큰 데이터에 대해서 결과를 Pass로 업데이트
df
```

```python
# df.loc['총합' > 400, '결과'] = 'Passs' # 오류 발생
# TypeError: '>' not supported between instances of 'str' and 'int'
df
```

## Cell 수정

```python
df.loc['4번','SW특기'] = 'Python' # 4번 학생의 SW특기 데이터를 Python으로 변경
df
```

## Column 순서 변경

```python
cols = list(df.columns)
cols
```

```python
df = df[[cols[-1]] + cols[:-1]] # 리스트 맨끝에 있는 항목을 아래로 가져오는 방법 : 슬라이싱 이용
# 리스트의 마지막 항목 + 리스트의 첫번째 항목부터 마지막 항목 앞까지 → 리스트간의 더하기
df
```
