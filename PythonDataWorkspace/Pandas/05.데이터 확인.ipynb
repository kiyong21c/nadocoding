```python
df.describe() # 계산이 가능한 데이터 Column에 대해서만
```

```python
df.info() # 각 Column별 타입, 메모리사용량, 컬럼 개수 등
```

## Series 확인

```python
df['키'].describe()
```

```python
df['키'].min() # 최소값
# df['키'].max() # 최대값
# df['키'].mean() # 평균값
# df['키'].sum() # 모든 키 시리즈를 더한 값
```
