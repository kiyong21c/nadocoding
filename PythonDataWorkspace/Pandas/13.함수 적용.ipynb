## 데이터에 함수 적용(apply)

```python
# 키 뒤에 cm 붙이는 역할
def add_cm(height):
    return str(height) + 'cm'

df['키'] = df['키'].apply(add_cm) # 키 데이터에 대해 add_cm함수를 호출한 결과 데이터를 반영
df
```

```python
def capitalize(lang):
    if pd.notnull(lang): # NaN이 아닌지(pandas에 존재하는 함수 활용)
        return lang.capitalize() # 파이썬에 원래 존재하는 함수(첫글자는 대문자 나머지 소문자), 위에서 직접 정의한 함수와는 다름
    return lang # NaN이면 그대로 NaN반환
df['SW특기'] = df['SW특기'].apply(capitalize)
df
```
