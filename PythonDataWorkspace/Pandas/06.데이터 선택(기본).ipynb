## Column 선택(정수 index)
데이터가 오른쪽으로 쌓이다 보면 컬럼명을 다 알기 어려울 수 있다

```python
df.columns
```

```python
df.columns[0]
```

```python
df.columns[2]
```

```python
df[df.columns[0]] # df['이름'] 과 동일한 동작 수행
```

```python
df[df.columns[-1]] # 맨 우측 column의 값을 가져옴
# 보통 최신 데이터를 우측에 순서대로 쌓기 때문에 자주 사용될 수 있음
```

## 슬라이싱

```python
df
```

```python
df['영어'][0:5] # 영어 column 중에서 0~4번째 값을 가져옴
```

```python
df[['이름','키']][:3] # 처음 3명의 이름과 키 정보를 가져옴
```

```python
df[3:] # 3번째부터 끝까지의 정보를 가져옴
# 위의 경우 컬럼의 순서라고 착각할 필요가 없다(기본적인 슬라이싱 문법이라고 볼 수 있음)
```
